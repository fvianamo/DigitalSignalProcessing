Esse programa tem como objetivo comparar o tempo de execução da transformada de fourier e a transformada rápida de fourier. O sinal de entrada para ambas as transformadas é um conjunto com 2048 amostras geradas aleatoriamente, com valores em ponto flutuante entre 0 e 100.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random

x = []
t = []

for i in range(0,2047):
    t.append(i)
    x.append(random.uniform(0.0,100.0))
    
plt.plot(t,x)
plt.axis([0, 2040, 0, 100])
plt.show()
```
